
Fuzzy Road Situation Model Optimization Routing (FRSMOR) in Vehicular Ad-hoc Network (VANET) 

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Input kepadatan kendaraan :

   Kepadatan Rendah (0,9 - 1,4 m)
   Kepadatan Sedang (1,25 - 1,8 m)
   Kepadatan Tinggi (1,6 - 3 m)
   
Input kecepatan rata-rata:

   Kecepatan Rendah (15 - 35 km/h)
   Kecepatan Sedang (40 - 50 km/h)
   Kecepatan Tinggi (45 - 80 km/h)
   
output 


   


```python
kepadatan = ctrl.Antecedent(np.arange(0,300,1),'kepadatan')# nilai satuan meter diubah ke centimeter
kepadatan_x = np.arange(0,300,1) 
```

# Fuzzifikasi Kepadatan Kendaraan

Proses untuk mengubah variabel non numerik menjadi variabel non fuzzy (numerik) menjadi variabel fuzzy (linguistik)

Untuk Kepadatan Kendaraan, fungsi keanggotaan dibagi menjadi 3 bagian :
                                                                        Tingkat Kepadatan
                                                                        Rendah=0,9m-1,4m
                                                                        Sedang=1,25m-1,8m
                                                                        Tinggi=1,6m-3m

# Dengan fungsi keanggotaan kepadatan kendaraan


```python
kepadatan['rendah'] = fuzz.trapmf(kepadatan.universe,[0,0,90,140]) #dari 0,9 meter - 1,4 meter
kepadatan['sedang'] = fuzz.trimf(kepadatan.universe,[70,125,180]) #dari 1,25 meter - 1,8 meter
kepadatan['tinggi'] = fuzz.trapmf(kepadatan.universe,[160,180,300,300]) #dari 1,6 meter - 3 meter
```

rentang kepadatan dibuah dari meter ke centimeter

```python
kepadatan_sedang = fuzz.trimf(kepadatan_x,[70,125,180])
fig2, (ax1) = plt.subplots(nrows=1,figsize=(6,6))
ax1.plot(kepadatan_x,kepadatan_sedang,'orange')
ax1.set_title('Kepadatan Sedang')
```

```python
kepadatan_tinggi = fuzz.trapmf(kepadatan_x,[160,180,300,300])
fig3, (ax2) = plt.subplots(nrows=1,figsize=(6,6))
ax2.plot(kepadatan_x,kepadatan_tinggi,'green')
ax2.set_title('Kepadatan Tinggi')
```

# Fuzzifikasi Kemacetan Kendaraan

Proses untuk mengubah variabel non numerik menjadi variabel non fuzzy (numerik) menjadi variabel fuzzy (linguistik)

```python
kemacetan = ctrl.Consequent(np.arange(0,60,1),'kemacetan')
kemacetan_x = np.arange(0,60,1) 
```

Untuk Kemacetan Kendaraan, fungsi keanggotaan dibagi menjadi 3 bagian : Tingkat Kemacetan Rendah=15-35 menit
Kemacetan Sedang=40-50 menit Kemacetan Tinggi=45-80 menit

# Dengan fungsi keanggotaan kemacetan kendaraan
\begin{equation}
  kemacetan['rendah'] = \mu KemacetanRendah(x)=\begin{cases}
    1, & \text{$0\leq x \leq 15$}.  \\ 
    \left.(30 - x) \,  \right/(30 - 15) , & \text{$15 \leq x \leq 30$}.\\
    0, & \text{$x\geq 30$}.
  \end{cases}
  \\
    kemacetan['sedang'] = \mu KemacetanSedang(x)=\begin{cases}
    0, & \text{$x\leq 25$ atau $x\geq40$}.\\
    \left.(x - 25)\, \right/(35 - 25) , & \text{$25 \leq x \leq 35$}.\\
    \left.(40 - x)\, \right/(40 - 35) , & \text{$35 \leq x \leq 40$}.\\
  \end{cases}
  \\
  kemacetan['tinggi'] = \mu KemacatanTinggi(x)=\begin{cases}
    0, & \text{$x \leq 35$}.\\
    \left.(x - 35) \,  \right/(50 - 35) , & \text{$35 \leq x \leq 40$}.\\
    1, & \text{$x\geq 40$}.
  \end{cases}
\end{equation}

```python
kemacetan['rendah'] = fuzz.trapmf(kemacetan.universe,[0,0,15,30]) #dari 15 - 30 menit
kemacetan['sedang'] = fuzz.trimf(kemacetan.universe,[25,35,40]) #dari 35 - 40 menit
kemacetan['tinggi'] = fuzz.trapmf(kemacetan.universe,[35,50,60,60]) #dari 50 - 60 menit
```

# PLOTTING MEMBERSHIP FUNCTION KEMACETAN

```python
kemacetan_rendah = fuzz.trapmf(kemacetan_x,[0,0,15,30])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(kemacetan_x,kemacetan_rendah,'blue')
ax0.set_title('kemacetan Rendah')
```

```python
kemacetan_sedang = fuzz.trimf(kemacetan_x,[25,35,40])
fig2, (ax1) = plt.subplots(nrows=1,figsize=(6,6))
ax1.plot(kemacetan_x,kemacetan_sedang,'orange')
ax1.set_title('kemacetan Sedang')
```

```python
kemacetan_tinggi = fuzz.trapmf(kemacetan_x,[35,50,60,60])
fig3, (ax2) = plt.subplots(nrows=1,figsize=(6,6))
ax2.plot(kemacetan_x,kemacetan_tinggi,'green')
ax2.set_title('kemacetan Tinggi')
```

```python
kemacetan.view()
```

# Inferencing
Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

Pada sistem ini berdasarkan jumlah fuzzifikasi dari variabel kepadatan yang berjumlah 3 dan fuzzifikasi dari variabel kecepatan yang berjumlah 3, maka rule yang dibuat berjumlah 3 x 3 = 9 rule

1. IF Kepadatan RENDAH dan Kecepatan RENDAH THEN Kemacetan RENDAH
2. IF Kepadatan RENDAH dan Kecepatan SEDANG THEN Kemacetan RENDAH
3. IF Kepadatan RENDAH dan Kecepatan TINGGI THEN Kemacetan SEDANG

4. IF Kepadatan SEDANG dan Kecepatan RENDAH THEN Kemacetan RENDAH
5. IF Kepadatan SEDANG dan Kecepatan SEDANG THEN Kemacetan SEDANG
6. IF Kepadatan SEDANG dan Kecepatan TINGGI THEN Kemacetan TINGGI

7. IF Kepadatan TINGGI dan Kecepatan RENDAH THEN Kemacetan SEDANG
8. IF Kepadatan TINGGI dan Kecepatan SEDANG THEN Kemacetan TINGGI
9. IF Kepadatan TINGGI dan Kecepatan TINGGI THEN Kemacetan TINGGI


```python
rule1 = ctrl.Rule(kepadatan['rendah'] & kecepatan['rendah'], kemacetan['rendah'])
rule2 = ctrl.Rule(kepadatan['rendah'] & kecepatan['sedang'], kemacetan['rendah'])
rule3 = ctrl.Rule(kepadatan['rendah'] & kecepatan['tinggi'], kemacetan['sedang'])

rule4 = ctrl.Rule(kepadatan['sedang'] & kecepatan['rendah'], kemacetan['rendah'])
rule5 = ctrl.Rule(kepadatan['sedang'] & kecepatan['sedang'], kemacetan['sedang'])
rule6 = ctrl.Rule(kepadatan['sedang'] & kecepatan['tinggi'], kemacetan['tinggi'])

rule7 = ctrl.Rule(kepadatan['tinggi'] & kecepatan['rendah'], kemacetan['sedang'])
rule8 = ctrl.Rule(kepadatan['tinggi'] & kecepatan['sedang'], kemacetan['tinggi'])
rule9 = ctrl.Rule(kepadatan['tinggi'] & kecepatan['tinggi'], kemacetan['tinggi'])
```

# RULE EVALUATION

```python
info_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

# DEFUZZIFIKASI

```python
infom = ctrl.ControlSystemSimulation(info_ctrl)
```

Input kepadatan kita masukan 300 dan kecepatan 80

```python
infom.input['kepadatan'] = 125
infom.input['kecepatan'] = 40
```

Melakukan perhitungan defuzzifikasi

```python
infom.compute()
```

Mencetak hasil output defuzzifikasi

```python
print(infom.output['kemacetan'])
```
