```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Input : 

1) Burst Time terdiri dari 3 values :                       
- low (0 - 8)

2) Nice Value terdiri dari 3 values :

# Fuzzifikasi Burst Time

Untuk Burst Time, fungsi keanggotaan terdiri dari 3 Value                      
- low (0 - 8)                                
- medium (5 - 13)(13 - 21)                   
- high (16 - 26)                             

Dengan Fungsi Keanggotaan :


```python
burst = ctrl.Antecedent(np.arange(0,26,1),'burst')
burst_x = np.arange(0,26,1) 
```

```python
# membership function
burst['low'] = fuzz.trimf(burst.universe,[0,0,8])
burst['medium'] = fuzz.trimf(burst.universe,[5,13,21])
burst['high'] = fuzz.trimf(burst.universe,[16,26,26])
```

PLOTTING MEMBERSHIP FUNCTION BURST TIME

```python
#BURST LOW
burst_low = fuzz.trimf(burst_x,[0,0,8])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(burst_x, burst_low,'blue')
ax0.set_title('Burst Time Low')
```

```python
#BURST MEDIUM
burst_medium = fuzz.trimf(burst_x,[5,13,21])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(burst_x, burst_medium,'red')
ax0.set_title('Burst Time Medium')
```

```python
#BURST HIGH
burst_high = fuzz.trimf(burst_x,[16,26,26])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(burst_x, burst_high,'green')
ax0.set_title('Burst Time High')
```

# Fuzzifikasi Dynamic Priority

```python
dynamic = ctrl.Consequent(np.arange(0,28,1),'Dynamic Priority')
dynamic_x = np.arange(0,28,1)
```

Hasil Dynamic Priority dari sistem fuzzy memiliki 5 fungsi keanggotaan yaitu :

- Very Low (0 - 5)
- Low (2 - 12)
- Average (9 - 19)
- High (16 - 26)
- Very High (23 - 28)

Dengan Fungsi keanggotaan :


```python
dynamic['very low'] = fuzz.trapmf(dynamic.universe,[0,0,5,5])
dynamic['low'] = fuzz.trimf(dynamic.universe,[2,7,12])
dynamic['average'] = fuzz.trimf(dynamic.universe,[9,14,19])
dynamic['high'] = fuzz.trimf(dynamic.universe,[16,21,26])
dynamic['very high'] = fuzz.trapmf(dynamic.universe,[23,23,28,28])
```

PLOTTING MEMBERSHIP FUNCTION DYNAMIC PRIORITY

```python
dynamic_vlow = fuzz.trapmf(dynamic_x,[0,0,5,5])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(dynamic_x,dynamic_vlow,'blue')
```

```python
dynamic_low = fuzz.trimf(dynamic_x,[2,7,12])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(dynamic_x,dynamic_low,'brown')
```

```python
dynamic_average = fuzz.trimf(dynamic_x,[9,14,19])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(dynamic_x,dynamic_average,'green')
```

```python
dynamic_high = fuzz.trimf(dynamic_x,[16,21,26])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(dynamic_x,dynamic_average,'orange')
```

```python
dynamic_vhigh = fuzz.trapmf(dynamic_x,[23,28,28,28])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(dynamic_x,dynamic_vhigh,'yellow')
```

```python
dynamic.view()
```

# Inferencing
Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

Pada sistem ini berdasarkan jumlah fuzzifikasi dari variabel Burst Time yang berjumlah 3 dan fuzzifikasi dari variabel Nice Value yang berjumlah 3, maka rule yang dibuat berjumlah 3 x 3 = 9 rule



```python
#RULE
rule1 = ctrl.Rule(burst['low'] & nice['low'], dynamic['very high'])
rule2 = ctrl.Rule(burst['low'] & nice['medium'], dynamic['high'])
rule3 = ctrl.Rule(burst['low'] & nice['high'], dynamic['high'])
rule4 = ctrl.Rule(burst['medium'] & nice['low'], dynamic['average'])
rule5 = ctrl.Rule(burst['medium'] & nice['medium'], dynamic['average'])
rule6 = ctrl.Rule(burst['medium'] & nice['high'], dynamic['low'])
rule7 = ctrl.Rule(burst['high'] & nice['low'], dynamic['low'])
rule8 = ctrl.Rule(burst['high'] & nice['medium'], dynamic['very low'])
rule9 = ctrl.Rule(burst['high'] & nice['high'], dynamic['very low'])
```

# RULE EVALUATION

```python
dp_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

# DEFUZZIFIKASI

```python
dpp = ctrl.ControlSystemSimulation(dp_ctrl)
```

Input Burst Time kita masukan   dan Nice Value 

```python
dpp.input['burst'] = 15
dpp.input['Nice Value'] = 27
```

Melakukan perhitungan defuzzifikasi

```python
dpp.compute()
```

Mencetak hasil output defuzzifikasi

```python
print(dpp.output['Dynamic Priority'])
```

Plotting fungsi keangggotaan Burst Time setelah dimasukkan input nilai pada Burst Time

```python
burst.view(sim=dpp)
```

Plotting fungsi keangggotaan Nice Value setelah dimasukkan input nilai pada Nice Value

```python
nice.view(sim=dpp)
```

Plotting fungsi keangggotaan Dynamic Priority setelah dilakukan perhitungan defuzzifikasi output nilai pada Dynamic Priority

```python
dynamic.view(sim=dpp)
```
