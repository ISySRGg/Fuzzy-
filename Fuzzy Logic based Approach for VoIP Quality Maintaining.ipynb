

 # Fuzzy Logic based Approach for VoIP Quality Maintaining


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Input :
1. Train Speed
   - Low (0, 0, 40, 70)
   - Medium (40, 70, 100)
   - High (70, 100, 120, 120)

   
2. Braking Force
   - Small (0, 0, 1, 2)
   - Medium (1, 2, 4)
   - Big (2, 4, 5, 5)
   

Output :
   Train Braking Distance
   - Jarak 300 M (0, 0, 300, 500)
   - Jarak 500 M (300, 500, 700)
   - Jarak 700 M (500, 700, 1000)
   - Jarak 1000 M (700, 1000, 1100)
   - Jarak 1200 M (1000, 1100, 1200, 1200)

Train Speed mempunyai rentang nilai dari 0 - 120 km/h

```python
train_speed = ctrl.Antecedent(np.arange(0,120,1),'train_speed')
train_speed_x = np.arange(0,120,1) #untuk plotting saja
```

# Fuzzifikasi Train Speed

Proses untuk mengubah variabel non numerik menjadi variabel non fuzzy (numerik) menjadi variabel fuzzy (linguistik)

Untuk train speed, fungsi keanggotaan dibagi menjadi 3 bagian :
$$ Train Speed $$
$$ Low = 0 - 70$$
$$ Medium = 40 - 100$$
$$ High = 70 - 120$$

```python
train_speed['low'] = fuzz.trapmf(train_speed.universe,[0,0,40,70]) #dari 0 - 70
train_speed['medium'] = fuzz.trimf(train_speed.universe,[40,70,100]) #dari 40 - 100
train_speed['high'] = fuzz.trapmf(train_speed.universe,[70,100,120,120]) #dari 70 - 120
```

# PLOTTING MEMBERSHIP FUNCTION TRAIN SPEED

Plotting fungsi keanggotaan $\mu low(x)$

```python
train_low = fuzz.trapmf(train_speed_x,[0,0,40,70])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,7))
ax0.plot(train_speed_x,train_low,'blue')
ax0.set_title('low')
```

Plotting fungsi keanggotaan $\mu medium(x)$

```python
train_medium = fuzz.trimf(train_speed_x,[40,70,100])
fig2, (ax1) = plt.subplots(nrows=1,figsize=(6,7))
ax1.plot(train_speed_x,train_medium,'green')
ax1.set_title('medium')
```

Plotting fungsi keanggotaan $\mu high(x)$

```python
train_high = fuzz.trapmf(train_speed_x,[70,100,120,120])
fig3, (ax2) = plt.subplots(nrows=1,figsize=(6,7))
ax2.plot(train_speed_x,train_high,'red')
ax2.set_title('high')
```

Train Braking Distance mempunyai rentang nilai dari 0 - 1200 m

```python
distance = ctrl.Consequent(np.arange(0,1200,1),'distance')
distance_x = np.arange(0,1200,1)
```

Jarak pengereman diberi nama variabel distance dengan rentang jarak 0 - 1200 m dibagi menjadi 5 fungsi keanggotaan yaitu :
- Jarak 300 M (0 m - 500 m)
- Jarak 500 M (300 m - 700 m)
- Jarak 700 M (500 m - 1000 m)
- Jarak 1000 M (700 m - 1100 m)
- Jarak 1200 M (1000 m - 1200 m)

Dengan fungsi keanggotaan sebagai berikut :

```python
distance['300'] = fuzz.trapmf(distance.universe,[0,0,300,500]) # dari 0 - 500 m
distance['500'] = fuzz.trimf(distance.universe,[300,500,700]) # dari 300 - 700 m
distance['700'] = fuzz.trimf(distance.universe,[500,700,1000]) #dari 500 - 1000 m
distance['1000'] = fuzz.trimf(distance.universe,[700,1000,1100]) #dari 700 - 1100 m
distance['1200'] = fuzz.trapmf(distance.universe,[1000,1100,1200,1200]) #dari 1000 - 1200 m
```

# Inferencing

Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF...THEN...

Pada Sistem ini berdasarkan jumlah fuzzifikasi dari variabel train speed yang berjumlah 3 dan fuzzifikasi dari variabel braking force yang berjumlah 3, maka rule yang akan dibuat berjumlah 3x3 = 9 rule

1. IF Train Speed LOW dan Braking Force SMALL THEN Train Braking Distance 500
2. IF Train Speed MEDIUM dan Braking Force SMALL THEN Train Braking Distance 500
3. IF Train Speed HIGH dan Braking Force SMALL THEN Train Braking Distance 1200
4. IF Train Speed LOW dan Braking Force MEDIUM THEN Train Braking Distance 300
5. IF Train Speed MEDIUM dan Braking Force MEDIUM THEN Train Braking Distance 300
6. IF Train Speed HIGH dan Braking Force MEDIUM THEN Train Braking Distance 1000
7. IF Train Speed LOW dan Braking Force BIG THEN Train Braking Distance 300
8. IF Train Speed MEDIUM dan Braking Force BIG THEN Train Braking Distance 700
9. IF Train Speed HIGH dan Braking Force BIG THEN Train Braking Distance 1000

```python
rule1 = ctrl.Rule(train_speed['low'] & brake_force['small'], distance['500'])
rule2 = ctrl.Rule(train_speed['medium'] & brake_force['small'], distance['500'])
rule3 = ctrl.Rule(train_speed['high'] & brake_force['small'], distance['1200'])

rule4 = ctrl.Rule(train_speed['low'] & brake_force['medium'], distance['300'])
rule5 = ctrl.Rule(train_speed['medium'] & brake_force['medium'], distance['300'])
rule6 = ctrl.Rule(train_speed['high'] & brake_force['medium'], distance['1000'])
                  
rule7 = ctrl.Rule(train_speed['low'] & brake_force['big'], distance['300'])
rule8 = ctrl.Rule(train_speed['medium'] & brake_force['big'], distance['700'])
rule9 = ctrl.Rule(train_speed['high'] & brake_force['big'], distance['1000'])
```

# Rule Evaluation

```python
stop_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

# DEFUZZIFIKASI

```python
stop = ctrl.ControlSystemSimulation(stop_ctrl)
```

input train speed = 80 dan braking force = 3

```python
stop.input['train_speed'] = 80
stop.input['brake_force'] = 3
```

Melakukan perhitungan defuzzifikasi

```python
stop.compute()
```

Mencetak hasil output defuzzifikasi

```python
print(stop.output['distance'])
```

Plotting fungsi keangggotaan train speed setelah dimasukkan input nilai pada train speed

```python
train_speed.view(sim=stop)
```

Plotting fungsi keangggotaan brake force setelah dimasukkan input nilai pada brake force

```python
brake_force.view(sim=stop)
```

Plotting fungsi keangggotaan distance setelah dilakukan perhitungan defuzzifikasi output nilai pada distance
