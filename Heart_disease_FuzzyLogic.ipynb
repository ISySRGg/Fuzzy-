## Heart Disease Diagnosis Based On Mediative Fuzzy Logic


Merancang sistem pakar berdasarkan pada jenis logika fuzzy untuk mendiagnosis penyakit jantung pada pasien. Ada 9 rule dengan 3 variabel sebagai input dan 3 variabel sebagai output

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Input :
    
1) Blood Pressure terdiri dari 3 values :
- low (111 - 134)
- medium (127 - 139)(139 - 153)
- high (142 - 157)(157 - 172)

2) Maximum Heart Rate terdiri dari 3 values :
- low (100 - 141)
- medium (111 - 152) (152 - 194)
- high (152-216)

3) Blood Sugar terdiri dari 1 values :
- True (105 - 120)


Output :

- Healthy (0-1) 
- Sick1 (1-2)
- Sick2 (3-4)

# 1) Fuzzifikasi Blood Pressure

Untuk Blood Pressure, fungsi keanggotaan terdiri dari 3 values :

- low (111 - 134)
- medium (127 - 139)(139 - 153)
- high (142 - 157)(157 - 172)

Dengan fungsi keanggotaanya :







```python
blood_pressure = ctrl.Antecedent(np.arange(0,400,1),'blood_pressure')
blood_x = np.arange(0,400,1) 
```

```python
# membership function
blood_pressure['low'] = fuzz.trapmf(blood_pressure.universe,[0,0,111,134])
blood_pressure['medium'] = fuzz.trimf(blood_pressure.universe,[127,139,153])
blood_pressure['high'] = fuzz.trapmf(blood_pressure.universe,[142,172,400,400])
```

### PLOTTING MEMBERSHIP FUNCTION BLOOD PRESSURE

```python
#BLOOD LOW
blood_low = fuzz.trapmf(blood_x,[0,0,111,134])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(blood_x, blood_low,'blue')
ax0.set_title('Blood Pressure Low')
```

```python
#BLOOD MEDIUM
blood_medium = fuzz.trimf(blood_x,[127,139,153])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(blood_x, blood_medium,'orange')
ax0.set_title('Blood Pressure Medium')
```

```python
#BLOOD HIGH
blood_high = fuzz.trapmf(blood_x,[142,172,400,400])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(blood_x, blood_high,'green')
ax0.set_title('Blood Pressure High')
```

```python
blood_pressure.view()
```

# 2) Fuzzifikasi Maximum Heart Rate

Untuk Maximum Heart Rate, fungsi keanggotaan terdiri dari 3 values :

- low (100 - 141)
- medium (111 - 152) (152 - 194)
- high (152-216)

Dengan fungsi keanggotaannya :








### PLOTTING MEMBERSHIP FUNCTION MAXIMUM HEART RATE

```python
max_heart = ctrl.Antecedent(np.arange(0,400,1),'max_heart')
max_heart_x = np.arange(0,400,1)
```

```python
# membership function
max_heart['low'] = fuzz.trapmf(max_heart.universe,[0,0,100,141])
max_heart['medium'] = fuzz.trimf(max_heart.universe,[111,152,194])
max_heart['high'] = fuzz.trapmf(max_heart.universe,[152,216,400,400])
```

```python
#MAX_HEART LOW
max_low = fuzz.trapmf(max_heart_x,[0,0,100,141])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(max_heart_x, max_low,'blue')
ax0.set_title('Maximum Heart Rate Low')
```

# 3) Fuzzifikasi Blood Sugar

Untuk Blood Sugar, fungsi keanggotaan terdiri dari 2 values

- True (105 - 120)
- False (121 - 160) 

Dengan fungsi keanggotaan :


    

### PLOTING MEMBERSHIP FUNCTION BLOOD SUGAR

```python
blood_sugar = ctrl.Antecedent(np.arange(100,250,1),'blood_sugar')
bloodsugar_x = np.arange(100,250,1)
```

```python
# membership function
blood_sugar['True'] = fuzz.trapmf(blood_sugar.universe,[105,120,250,250])
```

```python
#blood sugar True
blood_l = fuzz.trapmf(bloodsugar_x,[152,216,400,400])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(bloodsugar_x, blood_l,'blue')
ax0.set_title('Blood Sugar')
```

# 4) Fuzzifikasi result

Untuk Result/Output, fungsi keanggotaan terdiri dari 3 values :

- Healthy (0-1) 
- Sick1 (1-2)
- Sick2 (3-4)

Dengan fungsi keanggotaan sebagai berikut :






### PLOTING MEMBERSHIP FUNCTION RESULT

```python
result = ctrl.Consequent(np.arange(-1,5,1),'result')
result_x = np.arange(-1,5,1)
```

```python
result['Healthy'] = fuzz.trapmf(result.universe,[-1,-1,0.25,1])
result['Sick1'] = fuzz.trimf(result.universe,[1,1,2])
result['Sick2'] = fuzz.trimf(result.universe,[2,2,3])
```

```python
result_heal = fuzz.trapmf(result_x,[-1,-1,0.25,1])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(result_x,result_heal,'blue')
ax0.set_title('Healthy')
```

```python
result_sick1 = fuzz.trimf(result_x,[1,1,2])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(result_x,result_sick1,'orange')
ax0.set_title('Sick1')
```

```python
result_sick2 = fuzz.trimf(result_x,[2,2,3])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(result_x,result_sick2,'green')
ax0.set_title('Sick2')
```

```python
result.view() #plot fungsi keanggotaan 
```

# INFERCING

#### RULE BASE :





```python
rule1 = ctrl.Rule(blood_pressure['low'] & max_heart['low']  & blood_sugar['True'], result['Healthy'])
rule2 = ctrl.Rule(blood_pressure['low'] & max_heart['medium']  & blood_sugar['True'], result['Sick1'])
rule3 = ctrl.Rule(blood_pressure['low'] & max_heart['high']   & blood_sugar['True'], result['Healthy'])

rule4 = ctrl.Rule(blood_pressure['medium'] & max_heart['low']  & blood_sugar['True'], result['Healthy'])
rule5 = ctrl.Rule(blood_pressure['medium'] & max_heart['medium']  & blood_sugar['True'], result['Sick1'])
rule6 = ctrl.Rule(blood_pressure['medium'] & max_heart['high']  & blood_sugar['True'], result['Sick1'])

rule7 = ctrl.Rule(blood_pressure['high'] & max_heart['low']  & blood_sugar['True'], result['Sick1'])
rule8 = ctrl.Rule(blood_pressure['high'] & max_heart['medium']  & blood_sugar['True'], result['Sick2'])
rule9 = ctrl.Rule(blood_pressure['high'] & max_heart['high']  & blood_sugar['True'], result['Sick2'])
```

# RULE EVALUATION

```python
diagnosis_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

# DEFUZZIFIKASI

```python
diagnosis = ctrl.ControlSystemSimulation(diagnosis_ctrl)
```

## Contoh Kasus

Berikut adalah contoh masukkan inputan
- Blood Pressure = 130

```python
diagnosis.input['blood_pressure'] = 130
diagnosis.input['max_heart'] = 175
diagnosis.input['blood_sugar'] = 110
```
